```python
import numpy as np
import pandas as pd
import re
import datetime

data=pd.read_csv(r"Orders.csv")
```

```python
pd.options.display.max_columns = None
from IPython.display import display
display(data)
```

```python
data['Profit']=data['Profit'].replace('[\$,]', '', regex=True).astype(float)
data['Sales']=data['Sales'].replace('[\$,]', '', regex=True).astype(float)
```

```python
data['Order.Date']=pd.to_datetime(data['Order.Date'])
data['Ship.Date']=pd.to_datetime(data['Ship.Date'])
```

```python
import matplotlib as plt
%matplotlib inline
```

```python
data['month'] = pd.to_datetime(data['Order.Date']).dt.strftime('%m')
```

```python
plotsum=data.groupby(['month','Sub.Category'])['Quantity'].sum().unstack().plot()
#plotsum.xticks(np.arange(min(month), max(month)+1, 1.0))
plotsum.set_xticks(np.arange(1, 12, 1.0), minor=True)
plotsum.set_xticklabels(np.arange(1, 12, 1.0), fontdict=None, minor=False)
```

```python
for title, group in data.groupby(['month','Sub.Category']):
    group.plot(x='month', y=sum('Quantity'), title=title)
```
